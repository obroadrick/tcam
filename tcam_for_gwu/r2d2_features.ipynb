```python
first_room_id = list(first_hotel_features.keys())[i]
first_feature = first_hotel_features[first_room_id]

# second_room_id = list(second_hotel_features.keys())[i+1]
# second_feature = second_hotel_features[second_room_id]
```

```python
path_to_first_image = h2i[str(first_hotel_id)][i]['img'].replace('/lab/vislab/DATA/ActualFullTraffickCam/', '/pless_nfs/home/datasets/FullTraffickcam/')
# path_to_second_image = h2i[str(second_hotel_id)][i+1]['img'].replace('/lab/vislab/DATA/ActualFullTraffickCam/', '/pless_nfs/home/datasets/FullTraffickcam/')
```

```python
from matplotlib import pyplot as plt
from PIL import Image
from random import choice
import numpy as np
```

```python
%matplotlib inline

# shuffle(first_feature['xys'])
im = np.array(Image.open(path_to_first_image))

colors = ['r', 'g', 'b']
for point in first_feature['xys']:
    circle = plt.Circle((int(point[0]), int(point[1])), 1, color=choice(colors), alpha=0.25)
    plt.gca().add_patch(circle)
plt.gca().invert_yaxis()
# plt.imshow(im[::-1], origin='lower')
plt.imshow(im[::-1], origin='lower')

plt.show()
```
